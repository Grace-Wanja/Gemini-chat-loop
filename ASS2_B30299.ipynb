```python
#4.Conversation Example
prompt = """
You are a polite banking assistant.
User: Can you explain what mobile banking is?
Assistant:
"""
response = model.generate_content(prompt)
print(response.text)
```

**Answer:**
The **structure of the prompt** determines response style:
- A **short, direct prompt** → concise reply.
- A **context-rich prompt** (with examples or tone instructions) → more elaborate and detailed answer.


```python
#5.Few-Shot Prompting Example
prompt = """
Translate the following:
Example 1:
Input: 'Hello'
Output: 'Bonjour'

Example 2:
Input: 'Good morning'
Output:
"""
response = model.generate_content(prompt)
print(response.text)

```

**Answer:**
This example illustrates **few-shot prompting** because it teaches the model by example.  
If we reduce examples, the model becomes **less accurate** and may rely more on general patterns instead of specific translation styles.


```python
#6. Translation Example (English → Luganda)
prompt = "Translate this to Luganda: 'Education empowers the nation.'"
response = model.generate_content(prompt)
print(response.text)
```

**Answer:**
To translate to **Luganda**, specify:
- The **target language** in the prompt.
- The **tone** (formal/informal).
For example: “Translate to Luganda, using polite tone appropriate for academic settings.”


```python
#7. Temperature and Summarization
for t in [0.2, 0.7, 1.0]:
    model_temp = genai.GenerativeModel("models/gemini-2.5-flash", generation_config={"temperature": t})
    response = model_temp.generate_content("Summarize the benefits of renewable energy.")
    print(f"Temperature {t}:\n{response.text}\n")
```

**Answer:**
- **Low temperature (0.2)** → precise and factual summaries.  
- **Medium (0.7)** → balanced, natural tone.  
- **High (1.0)** → creative or varied wording, sometimes less consistent.  
Temperature controls **creativity vs. accuracy**.


```python
#8. Q&A Example with Constraint
prompt = "Answer this in under 20 words: What is Artificial Intelligence?"
response = model.generate_content(prompt)
print(response.text)
```

**Answer:**
When constrained to a word limit, the model focuses on **essential information**.  
If it cannot fit the response, it **simplifies language** or omits examples.

```python
#10. Creative Writing vs. Code Generation
creative_prompt = "Design an app for Harba Store that helps track inventory creatively."
code_prompt = "Write Python code for an inventory tracking app for Harba Store."

creative = model.generate_content(creative_prompt)
code = model.generate_content(code_prompt)

print("Creative Writing:\n", creative.text)
print("\nCode Generation:\n", code.text)

```

**Answer:**
Creative writing outputs **concepts and ideas**, focusing on design and experience.  
Code generation outputs **functional logic**, focusing on implementation details.  
Both complement each other but serve **different cognitive goals**.
